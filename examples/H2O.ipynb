```python
import sys
sys.path.append("..")
from funcs.pcohp_funcs import get_cheap_pcohp
from helpers.ase_helpers import get_atoms
from helpers.data_parsers import parse_data
calc_path = "H2O_calc"
data = parse_data(root=calc_path)
atoms = get_atoms(calc_path)
H_idcs = [i for i, el in enumerate(atoms.get_chemical_symbols()) if el == "H"]
O_idcs = [i for i, el in enumerate(atoms.get_chemical_symbols()) if el == "O"]
```

```python
import matplotlib.pyplot as plt
Erange, ctot = get_cheap_pcohp(H_idcs, O_idcs, calc_path, data=data)
plt.plot(Erange, ctot, label="H2/O pCOHP")
mu = data[-1]
plt.axvline(x=mu, c="red", label="$\mu$")
plt.ylabel("pCOHP (Hartree)")
plt.xlabel("E (Hartree)")
plt.legend()
```
