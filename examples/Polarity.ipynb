### The following is an expo on an ad-hoc approach to modifying the cross-projection matrix, P_uv, in order to isolate electron donating and accepting bonds. Results may not be reliable.
### Solid lines represent pcoop/cohp-type density, and dashed lines represent their integrations.
### Black lines represent traditional pcoop/cohp type analysis, while green and yellow lines signify analysis for electron acceptance for first and second atom collections, respectively.
### $\tilde{P}$ and $A$ signify the two current approaches to modifying $P$ for analyzing polarity.
### $\tilde{P}$ is a simple weighting of $P$, where are matrix elements $P_{\mu,\nu}$ are weighted by $\frac{|P_{\mu,\mu}|}{|P_{\mu,\mu}| + |P_{\nu,\nu}|}$
### $A$ comes from the observation that the electron acceptor atom group dominates for bonding interactions within the $\tilde{P}$ approach, and the electron donor atom group dominates for anti-bonding interactions. As such, we construct $A(\mu)$ as $\tilde{P}_{\mu}^{+} + \tilde{P}_{\nu}^{-}$, where superscripts +/- signify taking max/min($\tilde{P}_{\gamma}$, 0) for all tensor elements of $\tilde{P}_{\gamma}$ (positive elements of this matrix give bonding, negative elements give antibonding)
